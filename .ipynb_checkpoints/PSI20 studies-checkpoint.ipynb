```python
## Imports

from bokeh.models import HoverTool, ColumnDataSource, BoxZoomTool
from bokeh.models.ranges import Range1d
from bokeh.models.axes import LinearAxis
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
import seaborn as sns
import pandas as pd
import tradingeconomics as te
import statsmodels.api as sm
import statsmodels.tsa.stattools as ts
pd.set_option("display.max_rows",1000)
import numpy as np
%matplotlib inline
```

```python
## load data

tickers = ['san:sm', 'galp:pl', 'edp:pl', 'jmt:pl', 'edpr:pl', 'bcp:pl', 'nos:pl', 'bpi:pl', 'son:pl', 'altr:pl', 'rene:pl', 'cor:pl',
 'sem:pl', 'egl:pl', 'pti:pl', 'ibs:pl', 'sonc:pl', 'sct:pl', 'soni:pl', 'nba:pl',
 'slben:pl', 'scp:pl', 'fcp:pl']

data = {}

for ticker in tickers:
    try:
        df = pd.read_csv('C:\\Users\\utilizador\\PycharmProjects\\tss\\histdata\\stocks_psi_geral\\minimal\\{}_daily_minimal.csv'.format(ticker.replace(':','-')),
                               parse_dates=True, index_col=0)
        data[ticker] = df[df > '2017-01-01']
    except Exception as e:
        print(e)
```

```python
corr_matrix = pd.DataFrame(columns=tickers, index=tickers)
positive_corr = {}
negative_corr = {}
for t1 in tickers:
    positive_corr[t1] = {}
    negative_corr[t1] = {}
    for t2 in tickers:        
        corr_matrix[t1][t2] = round(float(data[t1][t1].corr(data[t2][t2])),4)
        if corr_matrix[t1][t2] != 1:
            if corr_matrix[t1][t2] > 0.6:
                positive_corr[t1][t2] = corr_matrix[t1][t2]
            if corr_matrix[t1][t2] < -0.6:     
                negative_corr[t1][t2] = corr_matrix[t1][t2]
```

```python
sns.set(rc={'figure.figsize':(30,30)})
sns.heatmap(corr_matrix.astype(float), annot=True)  
```

```python
def cointegration(df1, df2):
    comb_index = df1.index.union(df2.index)
    df1 = df1.reindex(comb_index, method='ffill')
    df2 = df2.reindex(comb_index, method='ffill')
    engle_granger = ts.coint(df1.values,df2.values)
    beta_hr = sm.OLS(df1.values, df2.values).fit().params[0]
    res = df1.values - beta_hr * df2.values    
    # Calculate and reports the CADF test on the residuals
    adf = ts.adfuller(res.flatten())
    return engle_granger, adf

coint_matrix_adf = pd.DataFrame(columns=tickers, index=tickers)
coint_matrix_engle = pd.DataFrame(columns=tickers, index=tickers)
cointegrated_adf = {}
cointegrated_engle = {}
for t1 in tickers:
    cointegrated_adf[t1] = {}   
    cointegrated_engle[t1] = {}
    for t2 in tickers:
        if t2 == t1:
            continue
        coint_matrix_adf[t1][t2] = round(float(cointegration(data[t1][t1],data[t2][t2])[1][0]),4)
        if coint_matrix_adf[t1][t2] > .2:
            continue
        coint_matrix_engle[t1][t2] = round(float(cointegration(data[t1][t1],data[t2][t2])[0][0]),4)    
        if coint_matrix_engle[t1][t2] < -2.0:
            cointegrated_engle[t1][t2] = coint_matrix_engle[t1][t2]
        if coint_matrix_adf[t1][t2] < -2.0:
            cointegrated_adf[t1][t2] = coint_matrix_adf[t1][t2]    
            
```
